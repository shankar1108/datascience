```python
# Create a dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3}

# Get value by key
value_a = my_dict.get('a')
print(value_a)  # Output: 1

# Add a key-value pair if key is not present
my_dict.setdefault('e', 5)
print(my_dict)  # Output: {'a': 1, 'b': 2, 'c': 3, 'e': 5}

# Remove and return value by key
value_b = my_dict.pop('b')
print(value_b)  # Output: 2
print(my_dict)  # Output: {'a': 1, 'c': 3, 'e': 5}

# Remove and return a key-value pair
key_value = my_dict.popitem()
print(key_value)  # Output: ('e', 5)
print(my_dict)    # Output: {'a': 1, 'c': 3}

# Update the dictionary
my_dict.update({'d': 4, 'e': 5})
print(my_dict)  # Output: {'a': 1, 'c': 3, 'd': 4, 'e': 5}

# Get keys
keys = my_dict.keys()
print(keys)  # Output: dict_keys(['a', 'c', 'd', 'e'])

# Get values
values = my_dict.values()
print(values)  # Output: dict_values([1, 3, 4, 5])

# Get items
items = my_dict.items()
print(items)  # Output: dict_items([('a', 1), ('c', 3), ('d', 4), ('e', 5)])

# Clear the dictionary
my_dict.clear()
print(my_dict)  # Output: {}

# Create a new dictionary from keys with a default value
new_dict = dict.fromkeys(['x', 'y', 'z'], 0)
print(new_dict)  # Output: {'x': 0, 'y': 0, 'z': 0}






# Copy a dictionary
copied_dict = new_dict.copy()
print(copied_dict)  # Output: {'x': 0, 'y': 0, 'z': 0}
```
